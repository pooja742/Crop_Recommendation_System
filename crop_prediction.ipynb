```python
df = pd.read_csv("D:\\Crop recommendation system\\Crop_recommendation.csv")
```

```python
x = df.drop('label', axis = 1)
y = df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, stratify = y, random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=2000, solver='saga')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# Create a pipeline: scales data before training
model = make_pipeline(
    StandardScaler(),
    LogisticRegression(
        solver='lbfgs',       # 'lbfgs' converges faster for small/medium datasets
        max_iter=5000,        # allow more iterations
        multi_class='auto',   # handles multi-class automatically
        random_state=42
    )
)

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
logistic_acc = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy of logistic regression is " + str(logistic_acc))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_2 = DecisionTreeClassifier(criterion='entropy',max_depth = 6, random_state = 2)
```

```python
model_2.fit(x_train, y_train)
```

```python
y_pred_2 = model_2.predict(x_test)
```

```python
decision_acc = accuracy_score(y_test, y_pred_2)
```

```python
print("Accuracy of decision  tree is " + str(decision_acc))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_3 = GaussianNB()
```

```python
model_3.fit(x_train, y_train)
```

```python
y_pred_3 = model_3.predict(x_test)
```

```python
naive_bayes_acc = accuracy_score(y_test, y_pred_3)
```

```python
print("Accuracy of naive_bayes is " + str(naive_bayes_acc))
```
